```python

csv_file = Path('taxi_trip_pricing.csv')
for parent in Path.cwd().parents:
    possible_path = parent / 'src' / 'taxipred' / 'data' / 'taxi_trip_pricing.csv'
    if possible_path.exists():
        csv_file = possible_path
        break

print(f"Läser från: {csv_file}")
df = pd.read_csv(csv_file)

print(f"\nData inläst")
print(f"Antal rader: {df.shape[0]:,}")
print(f"Antal kolumner: {df.shape[1]}")
```

```python
print("=== KOLUMNNAMN ===")
print(df.columns.tolist())

print("\n=== DATATYPER ===")
print(df.dtypes)

print("\n=== GRUNDLÄGGANDE INFO ===")
df.info()
```

```python
df_with_price = df[df['Trip_Price'].notna()].copy()
df_without_price = df[df['Trip_Price'].isna()].copy()

print(f"Rader med pris: {len(df_with_price)}")
print(f"Rader utan pris (för predictions): {len(df_without_price)}")
```

```python
df_clean = df_with_price.dropna()

print(f"Före: {len(df_with_price)} rader")
print(f"Efter: {len(df_clean)} rader")
print(f"Borttagna: {len(df_with_price) - len(df_clean)} rader")
```

```python
print("Kvarvarande nulls:")
print(df_clean.isnull().sum().sum())
```

```python
print(f"Min pris: {df_clean['Trip_Price'].min()}")
print(f"Max pris: {df_clean['Trip_Price'].max()}")
print(f"Medel: {df_clean['Trip_Price'].mean():.2f}")

# Ta bort outliers
df_clean = df_clean[df_clean['Trip_Price'] > 0]  # Tar bort negativa
df_clean = df_clean[df_clean['Trip_Price'] < 200]  # Inga extremt höga

print(f"\nEfter outlier removal: {len(df_clean)} rader")
```
